```python
import scipy.io
import numpy as np
import pandas as pd
import os

# Use relative path since we're running locally
mat_path = 'Data_structure_all_subs.mat'

# Check if file exists
if os.path.exists(mat_path):
    print(f"File found: {mat_path}")
else:
    print(f"File not found: {mat_path}")
    print(f"Current working directory: {os.getcwd()}")
    print(f"Files in current directory: {os.listdir('.')}")

# Load the .mat file
mat_data = scipy.io.loadmat(mat_path)

# Display all keys in the .mat file
print("\nKeys in the .mat file:")
for key in mat_data.keys():
    if not key.startswith('__'):
        print(f"  - {key}")

print("\n" + "="*50)
print("Detailed structure of each variable:")
print("="*50)

# Explore each variable
for key in mat_data.keys():
    if not key.startswith('__'):
        print(f"\n{key}:")
        print(f"  Type: {type(mat_data[key])}")
        print(f"  Shape: {mat_data[key].shape}")
        print(f"  Dtype: {mat_data[key].dtype}")
        
        # Show first few elements if it's an array
        if isinstance(mat_data[key], np.ndarray):
            if mat_data[key].size <= 20:
                print(f"  Full data: {mat_data[key]}")
            else:
                print(f"  First few elements: {mat_data[key].flat[:10]}")
                print(f"  ... (total size: {mat_data[key].size})")
```
